```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.dummy import DummyRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error
import pandas as pd
import numpy as np
df = pd.read_csv("../data/raw/insurance.csv")
```

```python
X = df.drop("charges", axis=1)
y = df["charges"]

print(X.head())
print(y.head())
```

```python
X_encoded = pd.get_dummies(X, drop_first=True)
print(X_encoded.head())
print(X_encoded.columns)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Make predictions on the test set

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
# Train metrics
train_mae = mean_absolute_error(y_train, y_train_pred)
train_mse = mean_squared_error(y_train, y_train_pred)
train_rmse = np.sqrt(train_mse)

# Test metrics
test_mae = mean_absolute_error(y_test, y_test_pred)
test_mse = mean_squared_error(y_test, y_test_pred)
test_rmse = np.sqrt(test_mse)

print("TRAIN METRICS")
print("MAE:", train_mae)
print("MSE:", train_mse)
print("RMSE:", train_rmse)

print("\nTEST METRICS")
print("MAE:", test_mae)
print("MSE:", test_mse)
print("RMSE:", test_rmse)
```

```python
# Inspecting the coefficients of the model
coefficients = pd.DataFrame(
    {"Features": X_train.columns, "Coefficients": model.coef_}
)
print(coefficients)
print("Intercept:", model.intercept_)
```

Baseline model: Linear Regression

Findings:
- Train and test metrics are close, so the model generalizes reasonably.
- smoker_yes is the strongest feature by far.
- RMSE is noticeably higher than MAE, suggesting some larger errors.

```python


dummy = DummyRegressor(strategy="mean")
dummy.fit(X_train, y_train)

dummy_test_pred = dummy.predict(X_test)

dummy_mae = mean_absolute_error(y_test, dummy_test_pred)
dummy_mse = mean_squared_error(y_test, dummy_test_pred)
dummy_rmse = np.sqrt(dummy_mse)

print("DUMMY BASELINE (mean predictor)")
print("MAE:", dummy_mae)
print("MSE:", dummy_mse)
print("RMSE:", dummy_rmse)
```

### Finding: baseline vs. dummy benchmark

The dummy benchmark predicts the **training mean** (~$13,346) for everyone,
ignoring all features. It is the "no-skill" floor any real model must beat.

| Metric | Dummy (mean) | Linear baseline | Improvement |
| ------ | -----------: | --------------: | ----------- |
| MAE    | $9,593       | $4,181          | **−56%** (typical error more than halved) |
| RMSE   | $12,466      | $5,796          | **−53%** |
| R²     | −0.0009      | 0.784           | from ~0 → explains **78%** of variance |

**Takeaway:** the Linear Regression baseline is clearly learning real signal —
it cuts typical error by more than half and explains ~78% of the variance in
`charges`, versus essentially none for the mean predictor. The dummy's R² of ≈ 0
confirms the train/test split is sound (no leakage inflating the floor). This
establishes a meaningful reference point: stronger models (Ridge/Lasso, tree
ensembles, log-target, interaction terms) should now be judged against the
linear baseline, not the dummy.


```python
# Log-transforming the target variable
# When you train linear regression on normal charges, the model is trying to predict the exact money.
# The problem is that a mistake on a very large charge dominates the model.
# By log-transforming the target variable, we can reduce the impact of large charges and make the model more robust.

y = np.log1p(df["charges"])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)


# Train metrics
train_mae = mean_absolute_error(y_train, y_train_pred)
train_mse = mean_squared_error(y_train, y_train_pred)
train_rmse = np.sqrt(train_mse)

# Test metrics
test_mae = mean_absolute_error(y_test, y_test_pred)
test_mse = mean_squared_error(y_test, y_test_pred)
test_rmse = np.sqrt(test_mse)

print("TRAIN METRICS")
print("MAE:", train_mae)
print("MSE:", train_mse)
print("RMSE:", train_rmse)

print("\nTEST METRICS")
print("MAE:", test_mae)
print("MSE:", test_mse)
print("RMSE:", test_rmse)
```

```python
y_train_actual = np.expm1(y_train)
y_test_actual = np.expm1(y_test)

y_train_pred_actual = np.expm1(y_train_pred)
y_test_pred_actual = np.expm1(y_test_pred)

train_mae_actual = mean_absolute_error(y_train_actual, y_train_pred_actual)
train_mse_actual = mean_squared_error(y_train_actual, y_train_pred_actual)
train_rmse_actual = np.sqrt(train_mse_actual)

test_mae_actual = mean_absolute_error(y_test_actual, y_test_pred_actual)
test_mse_actual = mean_squared_error(y_test_actual, y_test_pred_actual)
test_rmse_actual = np.sqrt(test_mse_actual)

print("TRAIN METRICS (original dollar scale)")
print("MAE:", train_mae_actual)
print("MSE:", train_mse_actual)
print("RMSE:", train_rmse_actual)

print("\nTEST METRICS (original dollar scale)")
print("MAE:", test_mae_actual)
print("MSE:", test_mse_actual)
print("RMSE:", test_rmse_actual)
```

The log-transformed linear regression improved MAE on the original dollar scale, suggesting better average-case prediction. However, RMSE became worse, which indicates that large prediction errors increased after back-transforming to the original scale. This suggests the log-transform helped the model fit typical charges better, but reduced its ability to handle some extreme high-charge cases. Therefore, both linear baselines are informative, but a more flexible nonlinear model is likely needed for stronger overall performance.
