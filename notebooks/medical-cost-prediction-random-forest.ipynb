```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df = pd.read_csv("../data/raw/insurance.csv")
```

```python
X = df.drop("charges", axis=1)
y = df["charges"]
```

```python
X_encoded = pd.get_dummies(X, drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
rf_model = RandomForestRegressor(n_estimators=200, random_state=42, n_jobs=-1)
rf_model.fit(X_train, y_train)
```

```python
y_train_pred = rf_model.predict(X_train)
y_test_pred = rf_model.predict(X_test)
```

```python
# Evaluate
train_mae = mean_absolute_error(y_train, y_train_pred)
train_mse = mean_squared_error(y_train, y_train_pred)
train_rmse = np.sqrt(train_mse)

test_mae = mean_absolute_error(y_test, y_test_pred)
test_mse = mean_squared_error(y_test, y_test_pred)
test_rmse = np.sqrt(test_mse)

print("RANDOM FOREST - TRAIN METRICS")
print("MAE:", train_mae)
print("MSE:", train_mse)
print("RMSE:", train_rmse)

print("\nRANDOM FOREST - TEST METRICS")
print("MAE:", test_mae)
print("MSE:", test_mse)
print("RMSE:", test_rmse)

# Feature importance
feature_importance = pd.DataFrame({
    "Feature": X_train.columns,
    "Importance": rf_model.feature_importances_
}).sort_values(by="Importance", ascending=False)

print("\nFeature Importances:")
print(feature_importance)
```

## Project A Summary and Findings

### Project Title

Medical Cost Prediction Using Regression Models

### Objective

The goal of Project A was to predict individual medical insurance charges using personal and demographic features from the insurance dataset. The target variable was `charges`, while the feature set included `age`, `sex`, `bmi`, `children`, `smoker`, and `region`. Since the target variable is numeric, the problem was correctly framed as a regression task.

### Data Understanding and Preprocessing

The dataset was first inspected to understand its structure, feature types, and quality. The target variable, `charges`, was identified as numeric, while `sex`, `smoker`, and `region` were categorical and required encoding before modeling. Missing values were checked and the dataset was found to be clean enough for baseline modeling. Initial inspection also suggested that `charges` was right-skewed, with some large values likely acting as high-cost outliers. This indicated that a simple linear model might struggle with extreme cases.

### Modeling Approach

A staged modeling strategy was used to build and compare progressively stronger baselines:

1. **Dummy Regressor (Mean Predictor)**
   This served as a naive benchmark and simply predicted the mean charge for every case.

2. **Linear Regression (Raw Target)**
   This was used as the first real baseline model because it is simple, interpretable, and appropriate for regression.

3. **Linear Regression with Log-Transformed Target**
   Since `charges` appeared right-skewed, a log transform was applied to test whether stabilizing the target distribution would improve performance.

4. **Random Forest Regressor**
   A more flexible nonlinear model was then trained to capture relationships and interactions that linear regression might miss.

### Model Results

#### 1. Dummy Regressor

* Test MAE: 9593.34
* Test RMSE: 12465.61

The dummy model performed poorly, confirming that the features contain meaningful signal and that simple average prediction is not sufficient.

#### 2. Linear Regression (Raw Target)

* Train MAE: 4208.23
* Train RMSE: 6105.55
* Test MAE: 4181.19
* Test RMSE: 5796.28

This model provided a strong interpretable baseline. Train and test errors were close, suggesting that the model generalized reasonably well and was not obviously overfitting. Coefficient analysis showed that `smoker_yes` was by far the strongest positive predictor, followed by `children`, `bmi`, and `age`.

#### 3. Linear Regression (Log-Transformed Target)

* Train MAE (original scale): 4270.28
* Train RMSE (original scale): 8389.73
* Test MAE (original scale): 3888.44
* Test RMSE (original scale): 7814.06

The log-transformed model improved test MAE, meaning it performed better on average-case predictions. However, RMSE became worse after transforming back to the original charge scale, suggesting poorer handling of some high-cost cases. This indicates that the log transform improved typical-case fit but increased larger prediction errors.

#### 4. Random Forest Regressor

* Train MAE: 1048.58
* Train RMSE: 1893.80
* Test MAE: 2557.99
* Test RMSE: 4584.12

The random forest regressor produced the best overall predictive performance. It reduced both MAE and RMSE substantially compared with the linear models, showing that the dataset contains nonlinear relationships that tree-based methods can capture more effectively. However, the noticeable gap between train and test performance suggests some degree of overfitting.

### Feature Findings

Feature importance from the random forest model showed the following ranking:

* `smoker_yes` — dominant feature
* `bmi`
* `age`
* `children`
* `sex_male` and region variables — relatively minor contributions

This confirms that smoking status is the strongest driver of medical charges in the dataset, with BMI and age also contributing meaningfully.

### Final Interpretation

Project A showed that a simple interpretable linear regression model can provide a solid baseline and generalize reasonably well, but it is limited by the linear assumptions of the model. The log-transformed linear regression improved average prediction error but performed worse on larger errors after returning to the original scale. The random forest regressor delivered the strongest test performance overall, indicating that nonlinear patterns and feature interactions are important in predicting insurance charges.

### Final Conclusion

The best-performing model for Project A was the **Random Forest Regressor**, while **Linear Regression** remained the best interpretable baseline. The project demonstrates the importance of benchmarking, comparing model families, and balancing predictive strength with interpretability.

### Suggested Next Steps

* Tune random forest hyperparameters to reduce overfitting
* Inspect residuals for difficult high-charge cases
* Test interaction features such as smoker × bmi
* Compare with additional models such as gradient boosting


```python
# Reduce overfitting in your random forest

rf_model = RandomForestRegressor(
    n_estimators=300,
    max_depth=8,
    min_samples_split=10,
    min_samples_leaf=4,
    max_features="sqrt",
    random_state=42,
    n_jobs=-1
)

rf_model.fit(X_train, y_train)

y_train_pred = rf_model.predict(X_train)
y_test_pred = rf_model.predict(X_test)

train_mae = mean_absolute_error(y_train, y_train_pred)
train_mse = mean_squared_error(y_train, y_train_pred)
train_rmse = np.sqrt(train_mse)

test_mae = mean_absolute_error(y_test, y_test_pred)
test_mse = mean_squared_error(y_test, y_test_pred)
test_rmse = np.sqrt(test_mse)

print("TUNED RANDOM FOREST - TRAIN METRICS")
print("MAE:", train_mae)
print("MSE:", train_mse)
print("RMSE:", train_rmse)

print("\nTUNED RANDOM FOREST - TEST METRICS")
print("MAE:", test_mae)
print("MSE:", test_mse)
print("RMSE:", test_rmse)
```

## Persist the model

Save the fitted model to `models/` with `joblib` so it can be reloaded without
retraining. A small `.meta.json` sidecar records the sklearn version, feature
order, and target transform — these guard against silent mismatches at load time.

```python
# Persist the fitted model so it can be reused without retraining.
# NOTE: models/ is git-ignored by design (see .gitignore) — this saves to local disk only.
# NOTE: `rf_model` currently holds the TUNED forest (lower overfit, slightly weaker
# test error). Re-run the untuned cell first if you want to save that variant instead.
import joblib, json, sklearn
from pathlib import Path

MODEL_DIR = Path("../models")
MODEL_DIR.mkdir(exist_ok=True)

model_path = MODEL_DIR / "insurance-random-forest-v1.joblib"
joblib.dump(rf_model, model_path)

meta = {
    "sklearn": sklearn.__version__,
    "features": list(X_train.columns),
    "target_transform": None,  # trained on raw dollars
}
json.dump(meta, open(model_path.with_suffix(".meta.json"), "w"), indent=2)

print("Saved:", model_path)
print("Reload : m = joblib.load('../models/insurance-random-forest-v1.joblib')")
print("Predict: dollars = m.predict(X_new)")
```
