```python
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/raw/insurance.csv")
X = df.drop("charges", axis=1)
y = df["charges"]

X_encoded = pd.get_dummies(X, drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
hgb_model = HistGradientBoostingRegressor(
    max_depth=4,
    learning_rate=0.05,
    max_iter=300,
    random_state=42
)
```

```python
hgb_model.fit(X_train, y_train)
```

```python
# Predict
y_train_pred = hgb_model.predict(X_train)
y_test_pred = hgb_model.predict(X_test)

# Evaluate
train_mae = mean_absolute_error(y_train, y_train_pred)
train_mse = mean_squared_error(y_train, y_train_pred)
train_rmse = np.sqrt(train_mse)

test_mae = mean_absolute_error(y_test, y_test_pred)
test_mse = mean_squared_error(y_test, y_test_pred)
test_rmse = np.sqrt(test_mse)

print("HIST GRADIENT BOOSTING - TRAIN METRICS")
print("MAE:", train_mae)
print("MSE:", train_mse)
print("RMSE:", train_rmse)

print("\nHIST GRADIENT BOOSTING - TEST METRICS")
print("MAE:", test_mae)
print("MSE:", test_mse)
print("RMSE:", test_rmse)
```

## Project A Summary and Findings

### Project Title

Medical Cost Prediction Using Regression Models

### Objective

The aim of Project A was to predict individual medical insurance charges using demographic and health-related features from the insurance dataset. The target variable was `charges`, while the feature columns were `age`, `sex`, `bmi`, `children`, `smoker`, and `region`. Since the target variable is numeric, the task was framed as a regression problem.

### Data Understanding and Preparation

The dataset was inspected to understand its structure, feature types, and overall quality. The target variable `charges` was confirmed to be numeric, while `sex`, `smoker`, and `region` were categorical and required encoding before modeling. Missing values were checked and the dataset was suitable for baseline modeling. Initial exploration suggested that `charges` was right-skewed, with a few very high values likely contributing to larger prediction errors.

### Modeling Strategy

A staged modeling approach was used in order to compare simple, interpretable, and more flexible models.

#### 1. Dummy Regressor

A naive benchmark was created using a dummy regressor that always predicted the mean insurance charge. This served as a reference point to confirm whether the real models were actually learning meaningful patterns.

#### 2. Linear Regression on Raw Target

Linear regression was used as the first real baseline model because it is simple, interpretable, and appropriate for regression tasks. This model provided a useful benchmark for understanding feature effects and generalization.

#### 3. Linear Regression on Log-Transformed Target

Because the target variable appeared right-skewed, a log transformation was applied to test whether stabilizing the target distribution would improve performance.

#### 4. Random Forest Regressor

A random forest regressor was then trained to capture nonlinear relationships and interactions that linear regression could not model effectively.

#### 5. HistGradientBoostingRegressor

Finally, a HistGradientBoostingRegressor was trained as a stronger tree-based model for tabular data. This became the best-performing model in the project.

### Model Results

#### Dummy Regressor

* Test MAE: 9593.34
* Test RMSE: 12465.61

This benchmark performed very poorly, showing that simply predicting the mean charge is not sufficient and that the features contain real predictive signal.

#### Linear Regression (Raw Target)

* Train MAE: 4208.23
* Train RMSE: 6105.55
* Test MAE: 4181.19
* Test RMSE: 5796.28

This model provided a strong and interpretable baseline. Train and test performance were close, suggesting reasonable generalization and no obvious overfitting. Coefficient analysis showed that `smoker_yes` was by far the strongest positive predictor, followed by `children`, `bmi`, and `age`.

#### Linear Regression (Log-Transformed Target)

* Train MAE (original scale): 4270.28
* Train RMSE (original scale): 8389.73
* Test MAE (original scale): 3888.44
* Test RMSE (original scale): 7814.06

The log-transformed model improved MAE on the original scale, which suggests better average-case prediction. However, RMSE became worse after transforming predictions back to the original dollar scale, indicating that larger errors increased. This suggests the log transform helped on more typical cases but weakened performance on some extreme high-charge observations.

#### Random Forest Regressor

* Train MAE: 1048.58
* Train RMSE: 1893.80
* Test MAE: 2557.99
* Test RMSE: 4584.12

The random forest regressor produced a major improvement over linear regression, showing that the data contains nonlinear structure. However, the large gap between training and test performance indicated noticeable overfitting.

#### HistGradientBoostingRegressor

* Train MAE: 2054.31
* Train RMSE: 3636.65
* Test MAE: 2548.24
* Test RMSE: 4441.49

This model delivered the best overall performance. It achieved the lowest test MAE and RMSE of all models tested. Compared with the random forest, it also showed a smaller train–test gap, suggesting better generalization and less overfitting, while still maintaining the strongest predictive power.

### Feature Findings

Across the models, the most influential features were consistently:

* `smoker_yes`
* `bmi`
* `age`
* `children`

Smoking status was the strongest driver of predicted insurance charges by a large margin. BMI and age also contributed meaningfully, while sex and region had relatively smaller effects.

### Interpretation of Overfitting

The final winning model, HistGradientBoostingRegressor, still showed some train–test gap, so mild overfitting is still present. However, this overfitting is much less severe than in the random forest model, and the model still achieved the best test performance. Therefore, the final result reflects a good trade-off between predictive strength and generalization rather than a perfectly overfit-free model.

### Final Conclusion

Project A demonstrated that simple models such as linear regression can provide a valuable interpretable baseline, but more flexible nonlinear models are needed to achieve stronger predictive performance on this dataset. Among all models tested, **HistGradientBoostingRegressor** was the best final model because it produced the lowest test MAE and RMSE while also generalizing better than the original random forest. The project also showed the value of using a naive benchmark, interpretable baseline, and stronger nonlinear models in sequence when building a practical machine learning solution.
